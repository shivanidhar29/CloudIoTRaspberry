```python
import pandas as pd
import math

A4=pd.DataFrame(columns=['Values', 'weight','Vendor1','Vendor2'])
print(A4)

```

```python
A4['Values']=['Cost','Scalability','SecurityFeatures','Customizationoption', 'MaintenanceCost','ServiceLevelOptions','In-houseSupportCapabilities','IntegrationWithCurrenServices','Data sources compatibilities']
A4['weight']=[20,8,15,7,12,8,16,15, 9]
A4['Vendor1']=[5, 6, 7, 8, 9, 4, 6, 3, 7]
A4['Vendor2']=[7, 4, 7, 3, 2, 8, 6, 6, 4]
print(A4)
```

```python
for index, row in A4.iterrows():
  A4['Vendor1'][index]=row['Vendor1']*row['weight']
  A4['Vendor2'][index]=row['Vendor2']*row['weight'] 

print(A4)
```

```python
totalVendor1=0
totalVendor2=0

for index, row in A4.iterrows():
  totalVendor1+=row['Vendor1']
  totalVendor2+=row['Vendor1']

print('The total for Vendor1 is {:.2f} and for Vnedor2 is {:.2f}'.format(totalVendor1, totalVendor2))
```

```python
import numpy as np
import matplotlib.pyplot as plt

axisLocations = np.linspace(start=0, stop=2 * np.pi, num=len(A4['Vendor1']), endpoint=False)
axisLocations = np.concatenate((axisLocations,[axisLocations[0]]))
Vendor1=A4['Vendor1']
Vendor2=A4['Vendor2']
ratingsAcircular=np.concatenate((Vendor1,[Vendor1[0]]))
ratingsBcircular=np.concatenate((Vendor2,[Vendor2[0]]))

plt.figure(figsize=(8,8))
plt.subplot(polar=True)
plt.plot(axisLocations, ratingsAcircular, 'o-', linewidth=2, label='In house development')
plt.plot(axisLocations, ratingsBcircular, 'd-', linewidth=2, label='Outsource')
plt.title('Project options comparision', size=20)
lines, labels=plt.thetagrids(np.degrees(axisLocations),labels=A4['Values'])
plt.legend()
plt.show()
```
