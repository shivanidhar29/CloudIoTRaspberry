```python
import pandas as pd
import plotly.graph_objects as go
```

```python
edgedf = pd.read_csv("/content/drive/MyDrive/IT496/fundedtrainingarea.csv")
edgedf
```

```python
allnodes = edgedf.Trainingarea.values.tolist() + edgedf.ProjectTeam.values.tolist()
source_indices = [allnodes.index(Trainingarea) for Trainingarea in edgedf.Trainingarea]
target_indices = [allnodes.index(ProjectTeam) for ProjectTeam in edgedf.ProjectTeam]
#color=['khaki','lightpink']
fig= go.Figure(data=[go.Sankey(
    node = dict( label = allnodes),
    link = dict( source = source_indices, target = target_indices, value = edgedf.membersInterested)
                
)])
fig.update_layout(title_text='Funded Training areas connected to Project Team')

fig.show()
```
