```python
import pandas as pd
import math

A4=pd.DataFrame(columns=['criteria', 'weight','ratingA','ratingB','scoreA', 'scoreB'])
print(A4)
```

```python
A4['criteria']=['risk','ROI','customerSatisfaction','feasibility', 'StrageticAlignment']
A4['weight']=[0.1,0.15,0.3,0.15,0.3]
A4['ratingA']=[4,1,2,1,5]
A4['ratingB']=[3,4,2,3,2]
print(A4)
```

```python
for index, row in A4.iterrows():
  A4['scoreA'][index]=row['ratingA']*row['weight']
  A4['scoreB'][index]=row['ratingB']*row['weight'] 

print(A4)
```

```python
totalScoreA=0
totalScoreB=0

for index, row in A4.iterrows():
  totalScoreA+=row['scoreA']
  totalScoreB+=row['scoreB']

print('The total score for A is {:.2f} and for B is {:.2f}'.format(totalScoreA, totalScoreB))
```

```python
import numpy as np
import matplotlib.pyplot as plt

axisLocations = np.linspace(start=0, stop=2 * np.pi, num=len(A4['ratingA']), endpoint=False)
axisLocations = np.concatenate((axisLocations,[axisLocations[0]]))
ratingA=A4['ratingA']
ratingB=A4['ratingB']
ratingsAcircular=np.concatenate((ratingA,[ratingA[0]]))
ratingsBcircular=np.concatenate((ratingB,[ratingB[0]]))

plt.figure(figsize=(8,8))
plt.subplot(polar=True)
plt.plot(axisLocations, ratingsAcircular, 'o-', linewidth=2, label='In house development')
plt.plot(axisLocations, ratingsBcircular, 'd-', linewidth=2, label='Outsource')
plt.title('Project options comparision', size=20)
lines, labels=plt.thetagrids(np.degrees(axisLocations),labels=A4['criteria'])
plt.legend()
plt.show()
```
