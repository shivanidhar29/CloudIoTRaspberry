```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv("/content/drive/MyDrive/Test_exam.csv")
df
```

```python
nodes = df.TrainingArea.values.tolist() + df.ProjectTeam.values.tolist()
source = [nodes.index(TrainingArea) for TrainingArea in df.TrainingArea]
target = [nodes.index(ProjectTeam) for ProjectTeam in df.ProjectTeam]
fig= go.Figure(data=[go.Sankey(
    node = dict( label = nodes),
    link = dict( source = source, target = target, value = df.Trainings)
                
)])
fig.show()
```
